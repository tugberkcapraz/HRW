```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import MaxNLocator
import os

# Set up the plot style
sns.set_theme(style="whitegrid")
sns.set_palette("deep")

# Create the output directory if it doesn't exist
output_dir = 'webapp/plots'
os.makedirs(output_dir, exist_ok=True)

# iterate over the accusation categories and bar plot the country counts
for category in df["long_term_consequence_category"].unique():
    country_counts = df[df["long_term_consequence_category"] == category]["Country"].value_counts()
    # Filter countries with counts greater than 1
    filtered_counts = country_counts[country_counts > 1]
    
    if len(filtered_counts) > 1:  # Only plot if there's more than one country
        fig, ax = plt.subplots(figsize=(12, 6))
        
        # Create horizontal bar plot
        bars = sns.barplot(x=filtered_counts.values, y=filtered_counts.index, ax=ax, palette="husl")
        
        # Customize the plot
        ax.set_title(f"Long Term Risk of {category}", fontsize=16, fontweight='bold', pad=20)
        ax.set_ylabel("Country", fontsize=12, labelpad=10)
        
        # Remove top and right spines
        sns.despine()
        
        # Customize ticks
        ax.xaxis.set_major_locator(MaxNLocator(integer=True))
        ax.tick_params(axis='both', which='major', labelsize=10)
        
        # Add value labels to the end of each bar
        for i, v in enumerate(filtered_counts.values):
            ax.text(v + 0.1, i, str(v), va='center', fontsize=10, fontweight='bold')
        
        # Adjust layout
        plt.tight_layout()
        
        # Center the x-label and data source text
        fig.text(0.5, 0.01, "Number of Occurrences", ha='center', va='bottom', fontsize=12)
        fig.text(0.5, -0.02, "Data Source: Human Rights Watch", ha='center', va='bottom', fontsize=10, style='italic')
        
        # Save the plot
        plt.savefig(os.path.join(output_dir, f"risk_of_{category.replace(' ', '_')}.png"), bbox_inches='tight')
        
        plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a contingency table
contingency_table = pd.crosstab(df['accusation_category'], df['long_term_consequence_category'])

# Set up the plot style
plt.figure(figsize=(12, 8))
sns.set_theme(style="whitegrid")

# Create a heatmap
heatmap = sns.heatmap(contingency_table, annot=True, fmt='d', cmap='Blues', cbar=True)

# Customize the plot
plt.title("Correlation between Accusation Categories and Long Term Consequence Categories", fontsize=16, fontweight='bold')
plt.xlabel("Long Term Consequence Category", fontsize=12)
plt.ylabel("Accusation Category", fontsize=12)
# save plot
plt.savefig('webapp/plots/correlation_between_accusation_and_long_term_consequence.png', bbox_inches='tight')
plt.tight_layout()
plt.show()
```
